```python
import matplotlib.pyplot as plt
from apyori import apriori
print(myntra)
```

```python
records=[]
for i in range(1,12491):
  records.append([str(myntra.values[i,j]) for j in range(0,8)])
```

```python
print(type(records))
```

```python
association_rules=apriori(records,min_support=0.001,min_confidence=0.3,min_lift=3, min_length=2)
association_results=list(association_rules)
print(association_results)
```

```python
print("there are {} relation derived.".format(len(association_results)))
```

```python
for i in range(0,len(association_results)):
  print(association_results[i][0])
```

```python
for item in association_results:
  pair=item[0]
  items=[x for x in pair]
  print("Rule: "+items[0]+"->"+items[1])
  print("Support: "+str(item[1]))
  print("Confidence: "+str(item[2][0][2]))
  print("Lift: "+str(item[2][0][3]))
  print("========================================")
```

**K-Means Clustering**

```python
myntra2.reset_index()
```

```python
myntra_kmeans=pd.get_dummies(myntra2,drop_first=True)
myntra_kmeans.head()
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaled_of_kmeans=scaler.fit_transform(myntra_kmeans)
```

```python
from sklearn.cluster import KMeans
kmeans_model=KMeans(n_clusters=7)
clusters=kmeans_model.fit_predict(myntra_kmeans)
myntra_kmeans.insert(myntra_kmeans.columns.get_loc("Price (INR)"),"Cluster",clusters)
myntra_kmeans
```

```python
myntra_kmeans["Cluster"].value_counts()
```

```python
myntra_kmeans.Cluster.unique()
```

```python
ssd=[]
for k in range(2,9):
  kmeans_model=KMeans(n_clusters=k)
  kmeans_model.fit(myntra_kmeans)
  ssd.append(kmeans_model.inertia_)
plt.figure(figsize=(6,4),dpi=100)
plt.plot(range(2,9),ssd,color='purple',marker='o')
plt.xlabel("number of clusters(K)")
plt.ylabel("SSD for X")
plt.show()
```

# DECISION TREE 

```python
myntra4 = myntra.drop(columns = ['ProductName','ProductBrand','Description'])
myntra4.info()
```

```python
myntra5 = myntra4.reindex(columns=['ProductID', 'Price (INR)', 'NumImages','PrimaryColor','Gender'])
myntra5.head()
```

```python
X=myntra5.iloc[:,:-1] 
feature_cols=['ProductID', 'Price (INR)', 'NumImages','PrimaryColor']
Y=myntra5.iloc[:,4].values
```

```python
from sklearn.preprocessing import LabelEncoder
labelencoder_X=LabelEncoder()
X=X.apply(LabelEncoder().fit_transform)
```

```python
from sklearn.tree import DecisionTreeClassifier
regressor = DecisionTreeClassifier(criterion='gini',max_depth=4,splitter='best')
X
regressor.fit(X,Y)

X_in=np.array([2,0,1,1])
y_pred=regressor.predict([X_in])
print("Prediction: ",y_pred)
```

```python
X_inn=X_in.reshape(1,-1)
```

```python
X
```

```python
from six import StringIO
from IPython.display import Image
from sklearn.tree import export_graphviz
import pydotplus

dot_data=StringIO()
feature_cols=['ProductID', 'Price (INR)', 'NumImages','PrimaryColor']
export_graphviz(regressor,out_file=dot_data, filled=True,rounded=True,special_characters=True, feature_names = feature_cols)
graph=pydotplus.graph_from_dot_data(dot_data.getvalue())
graph.write_png("Decisiontree.png")
Image(graph.create_png())
```

```python
accuracy = regressor.score(X_inn, y_pred)
print(f'The accuracy of the decision tree classifier is: {accuracy:.2f}')
```
